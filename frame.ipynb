```python
import cv2

def extract_frames(video_path, output_path):
    # 打開影片文件
    cap = cv2.VideoCapture(video_path)
    
    # 檢查影片是否成功打開
    if not cap.isOpened():
        print("Error: Unable to open video file.")
        return
    
    # 計算影片的幀數
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # 遍歷每個帧並提取圖像
    for frame_index in range(total_frames):
        # 讀取下一幀
        ret, frame = cap.read()
        
        # 確保成功讀取幀
        if ret:
            # 構造輸出文件名
            output_file = output_path + "/frame_" + str(frame_index).zfill(5) + ".jpg"
            
            # 保存圖像
            cv2.imwrite(output_file, frame)
            print("Frame {} extracted and saved as {}".format(frame_index, output_file))
        else:
            print("Error: Unable to read frame {} from video.".format(frame_index))
    
    # 釋放資源
    cap.release()
```

```python
user_name = input("輸入資料夾名稱：")
test_time = input("輸入測試時間：")
```

```python
import os
# 輸入影片路徑和輸出目錄
video_path = f"video/{user_name}/{test_time}.mp4"
output_directory = video_path[0:-4]
print(output_directory)
os.makedirs(output_directory, exist_ok=True)
extract_frames(video_path, output_directory)
```
